1)	Utilizando imagem relogio.tif e faça uma reamostragem de seus pixels para 300, 150 e 72 dpi. Sabe-se que a imagem original possui 1250 dpi. Fazer os cálculos por meio do seu algortimo. Não usar o opencv para a conversão

```python
# type your code here
# Python code to read image
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("relogio.tif", cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray', vmin=0, vmax=255);
plt.show(block=False)
plt.close()

print('Tamanho da imagem : ',img.shape)
# convertendo para 300dpi

height = img.shape[0]
width = img.shape[1]

def rescale_image(height, width, dpi):
  scale = dpi / 1250
  new_width = int(width * scale)
  new_height = int(height * scale)

  new_image = np.zeros((new_height, new_width), dtype=np.uint8)
  x_scale = width / new_width
  y_scale = height / new_height

  for x in range(new_width):
    for y in range(new_height):
      x_coord = int(x * x_scale)
      y_coord = int(y * y_scale)
      new_image[y, x] = img[y_coord, x_coord]
  
  return new_image

# Para 300 dpi:
image_300 = rescale_image(height, width, 300)
plt.imshow(image_300, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()

# Para 150 dpi:
image_150 = rescale_image(height, width, 150)
plt.imshow(image_150, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()

# Para 72 dpi:
image_72 = rescale_image(height, width, 72)
plt.imshow(image_72, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()

```

2)	Utilizando a imagem ctskull-256.tif, refaça a quantização de seus pixels utilizando de 7 a 1 bit(s) por pixel.  Fazer os cálculos por meio do seu algortimo. Não usar o opencv para a conversão

```python
def readGrayscaleImage(image):
    img = cv2.imread(image, 0)

    return img

def quantitizeImage(image, bit_level):
    new_image = np.array(image)
    shift = 8 - bit_level

    new_image >>= shift
    new_image <<= shift

    return new_image

def getQuantitizedImage(file, bit):
    img = readGrayscaleImage(file)
    outputImg = quantitizeImage(img, bit)
    plt.imshow(outputImg, cmap='gray', vmin=0, vmax=255)
    plt.show(block=False)
    plt.close()

getQuantitizedImage('./ctskull-256.tif', 1)
getQuantitizedImage('./ctskull-256.tif', 2)
getQuantitizedImage('./ctskull-256.tif', 3)
getQuantitizedImage('./ctskull-256.tif', 4)
getQuantitizedImage('./ctskull-256.tif', 5)
getQuantitizedImage('./ctskull-256.tif', 6)
getQuantitizedImage('./ctskull-256.tif', 7)
```
