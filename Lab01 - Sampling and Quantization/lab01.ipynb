1)	Utilizando imagem relogio.tif e faça uma reamostragem de seus pixels para 300, 150 e 72 dpi. Sabe-se que a imagem original possui 1250 dpi. Fazer os cálculos por meio do seu algortimo. Não usar o opencv para a conversão

```python
# type your code here
# Python code to read image
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.dpi'] = 200

img = cv2.imread("relogio.tif", cv2.IMREAD_GRAYSCALE)

#plt.imshow(img, cmap='gray', vmin=0, vmax=255);

#print('Tamanho da imagem : ',img.shape)
# convertendo para 300dpi

def calc_novo_tamanho(img, dpi, novo_dpi):
    """
    img: matriz da imagem,
    dpi: dpi original dessa imagem
    novo_dpi: qual o dpi desejado

    dpi = pixels/inch
    1250 = linhas/inch
    inch*1250 = linhas
    inch = linhas/1250
    inch = linhas/dpi

    novo_dpi = pixels/inch
    300 = pixels/inch
    300*inch = pixels
    pixels = 300*inch

    pixels = novo_dpi*(img.shape[]/dpi)
    """
    return int(novo_dpi*(img.shape[0]/dpi)), int(novo_dpi*(img.shape[1]/dpi))

resultados = [img]
testes = [300, 150, 72]
for dpi in testes:
    largura, altura = calc_novo_tamanho(img, 1250, dpi)
    teste = np.zeros((largura, altura))
    teste = img[::int(img.shape[0]/largura),::int(img.shape[1]/altura)]
    resultados.append(teste)

#plt.imshow(resultados[2], cmap='gray', vmin=0, vmax=255)
fig, axes = plt.subplots(1,4, layout="constrained")
axes[0].imshow(resultados[0], cmap='gray', vmin=0, vmax=255)
axes[1].imshow(resultados[1], cmap='gray', vmin=0, vmax=255)
axes[2].imshow(resultados[2], cmap='gray', vmin=0, vmax=255)
axes[3].imshow(resultados[3], cmap='gray', vmin=0, vmax=255)
```
