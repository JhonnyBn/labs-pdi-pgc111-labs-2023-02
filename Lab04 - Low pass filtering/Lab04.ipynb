1)	Aplique o filtro de Média a imagem abaixo para diferentes tamanhos de máscaras.
- 3x3; 7x7; 3x3 aplicado três vezes sequenciais.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import convolve2d

plt.rcParams['figure.dpi'] = 200

img = cv2.imread("sta2.png", cv2.IMREAD_GRAYSCALE)

mascara3 = np.ones((3,3))/9
mascara7 = np.ones((7,7))/49

img3 = convolve2d(img, mascara3)
img7 = convolve2d(img, mascara7)
img3x3 = convolve2d(img, mascara3)
img3x3 = convolve2d(img3x3, mascara3)
img3x3 = convolve2d(img3x3, mascara3)

fig, axes = plt.subplots(2,2, layout="constrained")
axes[0,0].imshow(img, cmap='gray', vmin=0, vmax=255)
axes[0,1].imshow(img3, cmap='gray', vmin=0, vmax=255)
axes[1,0].imshow(img7, cmap='gray', vmin=0, vmax=255)
axes[1,1].imshow(img3x3, cmap='gray', vmin=0, vmax=255)
```

2)	Aplique o filtro Gaussiano à imagem abaixo para diferentes valores de desvio padrão (σ) e tamanho de máscara.
- σ = 1.0 (kernel size 5×5).	σ = 2.0 (kernel size 9×9).	σ =  4.0 (kernel size 15×15).

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import convolve2d

plt.rcParams['figure.dpi'] = 200

img = cv2.imread("ben2.png", cv2.IMREAD_GRAYSCALE)

img1 = np.copy(img)
img1 = cv2.GaussianBlur(img1, (5,5), 1)
img2 = np.copy(img)
img2 = cv2.GaussianBlur(img2, (9,9), 2)
img4 = np.copy(img)
img4 = cv2.GaussianBlur(img4, (15,15), 4)

fig, axes = plt.subplots(2,2, layout="constrained")
axes[0,0].imshow(img, cmap='gray', vmin=0, vmax=255)
axes[0,1].imshow(img1, cmap='gray', vmin=0, vmax=255)
axes[1,0].imshow(img2, cmap='gray', vmin=0, vmax=255)
axes[1,1].imshow(img4, cmap='gray', vmin=0, vmax=255)
```

3)	Aplique o filtro de Mediana para a imagem abaixo, que apresenta um ruído *salt and pepper*
-	3×3 filter;	7×7 filter

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.dpi'] = 200

img = cv2.imread("sta2noi2.png", cv2.IMREAD_GRAYSCALE)

img3 = np.copy(img)
img3 = cv2.medianBlur(img3, 3)
img7 = np.copy(img)
img7 = cv2.medianBlur(img7, 7)

fig, axes = plt.subplots(1,3, layout="constrained")
axes[0].imshow(img, cmap='gray', vmin=0, vmax=255)
axes[1].imshow(img3, cmap='gray', vmin=0, vmax=255)
axes[2].imshow(img7, cmap='gray', vmin=0, vmax=255)
```
